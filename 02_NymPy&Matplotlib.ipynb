```python
#VECTOR INDEXTING

# a = [1,2,3,4,5], index starts with 0 to n-1

# a[2] = 2, a[2:4] = 3,4, a[-2:] = 4,5, a[::2] = 1,3,5

#this is called slicing of array index or fancy indexing
```

```python
#VECTOR OPS, numpy is extremely fast for vector ops
#ops are implemented in c++ to make it faster

#manipulation are done as odinary number, same size arrays else it is scaler
one = np.array([1,2])
two = np.array([1,2])

one + two + two
```

```python
print(one**two)
print(one*two)
print(one/two)
```

```python
#for scaler vals, scaler is broadcasted and operated to each element

print(one + 10)
print(two - 7)
print(one**2)
print(one//2)
```

```python
#DOT AND CROSS PRODUCT

print(np.dot(one, two))
print(np.cross(one, two))
```

```python
#max, argmax, min, argmin, sum, mean can also be done

np.mean(one)
np.sum(one)
np.max(one)
```

```python
x = np.arange(0, 10, 0.2)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
plt.plot(x, y_sin, label = 'sine', linewidth = 4)
plt.plot(x, y_cos, label = 'cos', linewidth = 2)

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("SINE AND COS")

plt.legend()
plt.show()
plt.close()
```

```python
#WORKING WITH MULTIPLE PLOTS

# subplot() command specifies numrows, numcols, plot_number(1 to row*col)
# subplot and subplots are DIFFERENT

x = np.arange(0, 10, 0.2)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```python
 plt.subplot(2,1,1) #row, col, refering to 1st subplot
 
 plt.plot(x, y_cos)
 plt.title("Cosine plot")
 
 plt.subplot(2,1,2) #row, col, refering to 2nd subplot
 plt.plot(x, y_sin)
 plt.title("Sine plot")
 
```
